```python
import os
import numpy as np
import torch
import glob
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as T
from torch.utils.data import DataLoader
from torch.autograd import Variable
import matplotlib.pyplot as plt
import pathlib
import matplotlib.image as mpimg

from torchvision.models import squeezenet1_1
import cv2
from io import open
from PIL import Image
```

```python
# checking for gpu and setting as gpu
if torch.cuda.is_available():
    device = torch.device('cuda:0')
elif torch.backends.mps.is_available():
    device = torch.device('mps')
else:
    device = torch.devic('cpu')
print(device)
```

```python
transform = T.Compose([
    T.Resize((150,150)),      # images already 150x150, but making sure they are
    T.RandomHorizontalFlip(), # def prob of 0.5 for variation in dataset
    T.ToTensor(),             # 0-255 to 0-1, numpy as tensors
    T.Normalize([0.5,0.5,0.5],
                [0.5,0.5,0.5])   # 0-1 to [-1,1]
])

train_path = './dataset/seg_train/seg_train/'
test_path = './dataset/seg_test/seg_test/'
pred_path = './dataset/seg_pred/seg_pred'
```

`ImageFolder` is a PyTorch dataset class specifically designed for loading image data when the data is organized in a folder structure where each subfolder represents a different class or category, and the images belonging to each class are stored within their respective subfolders.

```python
def split_data(train_path, test_path):
    train_loader = DataLoader(
        torchvision.datasets.ImageFolder(train_path, transform=transform),
        batch_size=32,
        shuffle=True,
        drop_last=True
    )
    test_loader = DataLoader(
        torchvision.datasets.ImageFolder(test_path, transform=transform),
        batch_size=32,
        shuffle=True,
        drop_last=True
    )
    return train_loader, test_loader

train_loader, test_loader = split_data(train_path, test_path)
```

The images in pytorch format for dataset = (256,3,150,150) where 256 is num of batches, 3 is num of channels, 150 as width, height of image.

The formula for size of image in the current layer = $(\frac{(Num of pixels in prev layer+2*padding-number of pixels in kernel in h)}{stride})+1$. 

This is the formula used to calculate the output of the convolution layer.

* nn.Conv2d(3,64,3,padding=1) where 3 is num of channels, 64 as num of feature maps, 3 as kernel size
     - conv1 output = np.floor(((150+2x1-3)/1)+1 = 150 {Shape: (256,64,150,150)}
     - pooling output = 150/2 = 75  {Shape: (256,64,75,75)}
* nn.Conv2d(75,128,3, padding=1) where 75 as output from prev layer, 128 as num of feature maps, 3 as kernel size, padding as 1
    - conv2 output = np.floor(((75+2x1-3)/1)+1) = 75 {Shape: (256,128,75,75)}
    - pooling output = 75/2 = 37 {Shape: (256,128,37,37)}
* nn.Conv2d(128,256,3, padding=1) where 128 as output from prev layer, 256 as num of feature maps, 3 as kernel size, padding as 1
    - conv3 output = np.floor(((37+2x1-3)/1)+1) = 37 {Shape: (256,256,75,75)}
    - pooling output = 37/2 = 18 {Shape: (256,256,18,18)}
* fc1 input =  output in prev * output in prev * num of feature maps in prev layer = 18x18x48

```python
def create_model():
    class cnn_net(nn.Module):
        def __init__(self, num_classes=6):
            super().__init__()
            self.conv1 = nn.Conv2d(3,15,3,padding=1)
            self.bnorm1 = nn.BatchNorm2d(15)
            
            self.conv2 = nn.Conv2d(15,30,3, padding=1)
            self.bnorm2 = nn.BatchNorm2d(30)
            
            self.conv3 = nn.Conv2d(30,60,3, padding=1)
            self.bnorm3 = nn.BatchNorm2d(60)
            
            self.fc1 = nn.Linear(18*18*60, 64)
            self.out = nn.Linear(64,num_classes)
            
        def forward(self,x):
            # first block: convolution -> maxpool -> batchnorm -> relu
            x = F.max_pool2d(self.conv1(x), 2)
            x = F.relu(self.bnorm1(x))
            
            x = F.max_pool2d(self.conv2(x), 2)
            x = F.relu(self.bnorm2(x))
            
            x = F.max_pool2d(self.conv3(x), 2)
            x = F.relu(self.bnorm3(x))
            
            # reshape for linear layer
            n_units = x.shape.numel()/x.shape[0]
            x = x.view(-1, int(n_units))
            
            x = F.relu(self.fc1(x))
            x = self.out(x)
            return x
    net = cnn_net().to(device)
    loss_func = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=0.001, weight_decay=0.0001)
    return net, loss_func, optimizer
```

```python
def train_model(net, loss_func, optimizer, train_loader, test_loader):
    num_epochs = 12
    
    train_losses = torch.zeros(num_epochs)
    test_losses = torch.zeros(num_epochs)
    train_acc = torch.zeros(num_epochs)
    test_acc = torch.zeros(num_epochs)
    
    for epochi in range(num_epochs):
        net.train()
        batch_acc = []
        batch_loss = []
        for x,y in train_loader:
            x = x.to(device)
            y = y.to(device)
            
            y_hat = net(x)
            loss = loss_func(y_hat, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            batch_loss.append(loss.item())
            batch_acc.append(torch.mean((torch.argmax(y_hat,axis=1) == y).float()).item())
        
        train_acc[epochi] = 100*np.mean(batch_acc)
        train_losses[epochi] = np.mean(batch_loss)
        
        batch_test_acc = []
        batch_test_loss = []
        net.eval()
        for x,y in test_loader:
            x = x.to(device)
            y = y.to(device)
            
            with torch.no_grad():
                y_hat = net(x)
                loss = loss_func(y_hat, y)
            
            batch_test_acc.append(torch.mean((torch.argmax(y_hat,axis=1) == y).float()).item())
            batch_test_loss.append(loss.item())
        test_acc[epochi] = 100*np.mean(batch_test_acc)
        test_losses[epochi] = np.mean(batch_test_loss)
        print(f"Epoch: {epochi+1}/{num_epochs}")

    return train_acc, test_acc, train_losses, test_losses
```

```python
net, loss_func, optimizer = create_model()
train_acc, test_acc, train_losses, test_losses = train_model(net, loss_func, optimizer, train_loader, test_loader)
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(train_losses, label="Train")
ax[0].plot(test_losses, label="Test")
ax[0].set(xlabel="Epochs", ylabel="Loss", title="Model Loss")
ax[0].legend()

ax[1].plot(train_acc, label="Train")
ax[1].plot(test_acc, label="Test")
ax[1].set(xlabel="Epochs", ylabel="Accuracy", title=f"Test Accuracy: {test_acc[-1]:.2f}")
ax[1].legend()
plt.show()
```

## Now that training is done, let's predict on the test set

```python
test_transform = T.Compose([
    T.Resize((150,150)),      # images already 150x150, but making sure they are
    T.ToTensor(),             # 0-255 to 0-1, numpy as tensors
    T.Normalize([0.5,0.5,0.5],
                [0.5,0.5,0.5])   # 0-1 to [-1,1]
])

def predict(img_path, test_transform, net, classes):
    net.eval()
    images = Image.open(img_path)
    img_tensor = test_transform(images).float()
    
    img_tensor = img_tensor.unsqueeze_(0)
    
    if torch.cuda.is_available():
        img_tensor = img_tensor.cuda()
    input = Variable(img_tensor) # convert tensor to variable
    
    with torch.no_grad():
        output = net(input)
    index = output.data.cpu().numpy().argmax()
    pred = classes[index]
    
    return pred
```

```python
pred_images_path = glob.glob(pred_path + '/*.jpg')
pred_dict = {} # save image name as key, and pred as value

for i in pred_images_path:
    pred_dict[i[i.rfind('/')+1:]] = predict(i, test_transform, net, classes)
```

```python
# Define a function to display images with predictions
def display_images_with_predictions(pred_dict, pred_images_dir, num_images=10):
    plt.figure(figsize=(12, 8))
    
    # Limit the number of images to display
    num_images = min(num_images, len(pred_dict))
    
    # Get the first num_images items from pred_dict
    items_to_display = list(pred_dict.items())[:num_images]
    print(items_to_display)
    for i, (image_name, prediction) in enumerate(items_to_display, start=1):
        # Construct the full image path
        image_path = os.path.join(pred_images_dir, image_name)
        
        # Normalize the path to use forward slashes to handle both Windows and Unix-style paths
        image_path = os.path.normpath(image_path)
        print(image_path)
        # Load and display the image
        img = mpimg.imread(image_path)
        plt.subplot(3, 3, i)  # Change the subplot layout as needed
        plt.imshow(img)
        plt.title(f"Prediction: {prediction}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Call the function to display the first 10 images with predictions
display_images_with_predictions(pred_dict, "./dataset/seg_pred", num_images=9)
```
